# Data Preprocessing
## Missing Values

#### missing values ni to'ldirishda nega fixed qiymat, median yoki dropna bilan to'ldirmadim sababi missing values bor ustunda tushub qolganlar soni 4900 tadan ziyod edi fixed qilsam bir xil to'ldirgani uchun data imbalance bo'lib qolishi mumkin , dropna qilsam ustunlar qatorlari 4000 row atrofiga tushub  qolar edi

```python
def missing_values(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=="object":
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(), inplace=True)
    return df
```

```python
missing_values(df)
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder                                   
def encoding(df):
    encoder=LabelEncoder()
    for col in df.columns:
        if df[col].dtype=="object":
            if df[col].nunique()<=2:
                dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df         
```

```python
df=encoding(df)
```

```python
df.info()
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
num_col=df.select_dtypes(include=["float64", "int64"]).columns.drop("Electric Range")
df[num_col]=minmax.fit_transform(df[num_col])
```
